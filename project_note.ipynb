```python
import pandas as pd
# Read data from file
df = pd.read_csv('adult.data.csv')
df
```

```python
# How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.
race_count = df['race'].value_counts()
race_count
```

```python
# What is the average age of men?
average_age_men = round(float(df[df['sex']=='Male']['age'].mean()),1)
average_age_men 
```

```python
# What is the percentage of people who have a Bachelor's degree?
percentage_bachelors = round(int(df[df['education']=='Bachelors']['education'].count())*100.0/float(df['education'].count()),1)
percentage_bachelors 
```

```python
# What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
# What percentage of people without advanced education make more than 50K?
# with and without `Bachelors`, `Masters`, or `Doctorate`
higher_education =float(df[(
        ((df['education']=='Bachelors')|
        (df['education']=='Masters') |
        (df['education']=='Doctorate')))]['education'].count())
higher_education
```

```python

lower_education =float(df[(
        ~((df['education']=='Bachelors')|
        (df['education']=='Masters') |
        (df['education']=='Doctorate')))]['education'].count())
lower_education
```

```python

# percentage with salary >50K
higher_education_rich = round(int(
    df[(
        ((df['education']=='Bachelors')|
        (df['education']=='Masters') |
        (df['education']=='Doctorate')) & (df['salary']=='>50K')
    )]
    ['education'].count())*100.0/float(df[(
        ((df['education']=='Bachelors')|
        (df['education']=='Masters') |
        (df['education']=='Doctorate')))]['education'].count()),1)
higher_education_rich
```

```python

lower_education_rich = round(int(
    df[(
        ~((df['education']=='Bachelors')|
        (df['education']=='Masters') |
        (df['education']=='Doctorate')) & (df['salary']=='>50K')
    )]
    ['education'].count())*100.0/float(df[(
        ~((df['education']=='Bachelors')|
        (df['education']=='Masters') |
        (df['education']=='Doctorate')))]['education'].count()),1)
lower_education
```

```python

# What is the minimum number of hours a person works per week (hours-per-week feature)?
min_work_hours = min(df['hours-per-week'])
min_work_hours
```

```python
# What percentage of the people who work the minimum number of hours per week have a salary of >50K?
num_min_workers = df[df['hours-per-week']==min_work_hours]
num_min_workers
```

```python
rich_percentage = int(num_min_workers[num_min_workers['salary']==">50K"]['salary'].count()*100.0/num_min_workers['salary'].count())
rich_percentage
```

```python

# What country has the highest percentage of people that earn >50K?
highest_earning_country =(df[df['salary']==">50K"]['native-country'].value_counts())
highest_earning_country.index[0]
```

```python

highest_earning_country_percentage = round(float(highest_earning_country[0]*100.0/highest_earning_country.sum()),1)
highest_earning_country_percentage
```

```python
# Identify the most popular occupation for those who earn >50K in India.
top_IN_occupation = df[(df['native-country']=='India')&( df['salary']==">50K")]['occupation'].value_counts()
top_IN_occupation.index[0]
```
